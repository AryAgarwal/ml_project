````python
import gentle

# Load the audio file
audio_file = 'audio.wav'

# Load the transcript text
with open("transcription.txt", "r") as f:
    transcript_text = f.read()

pipeline = gentle.resources.TranscriptionPipeline()

# Perform forced alignment
aligned_transcript = pipeline.transcribe(
    audio_file,
    transcript_text,
    word_align=True,  # Align at the word level
    scorer="scorers/gentle_scorer.scorer"  # Use the default scorer
)
````

````python


import os
import aeneas

# Load the audio and text
audio_file = "audio.wav"
text_file = "transcription.txt"

# Get the path to the English language model
models_path = os.path.join(aeneas.configuration.Configuration().models_path, "en")

# Perform forced alignment
align = aeneas.Aligner(models_path, audio_file, text_file)
aligned_transcript = align.transcripts[0]
````

````python
from transformers import BertTokenizer, BertModel
import numpy as np
from sklearn.cluster import KMeans

# Load the pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

# Split the transcript into sentences
sentences = [sent.text_content for sent in aligned_transcript.sentences]

# Obtain sentence embeddings using BERT
embeddings = []
for sentence in sentences:
    inputs = tokenizer(sentence, return_tensors="pt")
    outputs = model(**inputs)
    embeddings.append(outputs.last_hidden_state.mean(dim=1).squeeze().detach().numpy())

# Cluster the embeddings using K-Means
num_clusters = 15  # Adjust the number of clusters as needed
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(embeddings)

# Group sentences into semantic chunks based on cluster labels
semantic_chunks = []
for cluster_id in range(num_clusters):
    chunk_sentences = [sentences[i] for i, label in enumerate(cluster_labels) if label == cluster_id]
    chunk_text = " ".join(chunk_sentences)
    start_time = aligned_transcript.sentences[np.argmin([i for i, label in enumerate(cluster_labels) if label == cluster_id])].start
    end_time = aligned_transcript.sentences[np.argmax([i for i, label in enumerate(cluster_labels) if label == cluster_id])].end
    chunk_length = end_time - start_time
    if chunk_length > 15:  # Split long chunks into 15-second segments
        num_segments = int(np.ceil(chunk_length / 15))
        segment_duration = chunk_length / num_segments
        for i in range(num_segments):
            start = start_time + i * segment_duration
            end = min(start + segment_duration, end_time)
            semantic_chunks.append({
                "chunk_id": len(semantic_chunks) + 1,
                "chunk_length": end - start,
                "text": chunk_text[int(start):int(end)],
                "start_time": start,
                "end_time": end,
            })
    else:
        semantic_chunks.append({
            "chunk_id": len(semantic_chunks) + 1,
            "chunk_length": chunk_length,
            "text": chunk_text,
            "start_time": start_time,
            "end_time": end_time,
        })
````
